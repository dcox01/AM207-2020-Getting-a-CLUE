# Create Fake Loan Data

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
import seaborn as sns
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```
pay_num = 8000
default_num = 2000
```

```
#                   |FICO|  term|   PTI|    LTV|    age|
pay_mean = np.array([650,   6,      0.1,    1.05,   42])

pay_cov = np.array([[10000, -10,    -0.75,  -6,     140],
                    [-10,   1,      0.01,   0.048,  -0.7],
                    [-0.75, 0.01,   0.0025, 0.001,  -0.014],
                    [-6,    0.048,  0.001,  0.04,   -0.112],
                    [140,   -0.7,   -0.014, -0.112, 49]])

#                       |FICO|  term|   pti|      ltv|      age|
default_mean = np.array([550,   6.5,    0.12,     1.15,     39])

default_cov = np.array([[5625,    5.25,   -0.1875,  0.75,     45],
                        [5.25,    0.49,   0.0035,   0.007,    -0.21],
                        [-0.1875, 0.0035, 0.0025,   -0.0002,  0],
                        [0.75,    0.007,  -0.0002,  0.04,     -0.036],
                        [45,      -0.21,  0,        -0.036,   36]])
```

```
default_data = np.append(np.random.multivariate_normal(default_mean,default_cov,default_num),np.ones((default_num,1)),axis=1)
pay_data = np.append(np.random.multivariate_normal(pay_mean,pay_cov,pay_num),np.zeros((pay_num,1)), axis=1)
```

```
df = pd.DataFrame(np.vstack((default_data,pay_data)))
df.columns = ('fico','term','pti','ltv','age','default')
df = shuffle(df).reset_index(drop=True)
```

```
fig, ax = plt.subplots(1, 5, figsize=(25, 5))
sns.histplot(df.fico[df.default==1],color='red',alpha=0.4,ax=ax[0],label='default')
sns.histplot(df.fico[df.default==0],color='blue',alpha=0.4,ax=ax[0],label='pay')
ax[0].legend(loc='upper left')
sns.histplot(df.term[df.default==1],color='red',alpha=0.4,ax=ax[1],label='default')
sns.histplot(df.term[df.default==0],color='blue',alpha=0.4,ax=ax[1],label='pay')
ax[1].legend(loc='upper left')
sns.histplot(df.pti[df.default==1],color='red',alpha=0.4,ax=ax[2],label='default')
sns.histplot(df.pti[df.default==0],color='blue',alpha=0.4,ax=ax[2],label='pay')
ax[2].legend(loc='upper left')
sns.histplot(df.ltv[df.default==1],color='red',alpha=0.4,ax=ax[3],label='default')
sns.histplot(df.ltv[df.default==0],color='blue',alpha=0.4,ax=ax[3],label='pay')
ax[3].legend(loc='upper left')
sns.histplot(df.age[df.default==1],color='red',alpha=0.4,ax=ax[4],label='default')
sns.histplot(df.age[df.default==0],color='blue',alpha=0.4,ax=ax[4],label='pay')
ax[4].legend(loc='upper left')
plt.show()
```

```
#correlation of inputs and output
corr = df.corr().round(1)
sns.set(font_scale=0.8)
cmap = sns.diverging_palette(240, 10, n=9)
sns.heatmap(corr, annot=True, cmap=cmap,
        xticklabels=df.columns.values,
        yticklabels=df.columns.values)
plt.title('Feature Correlation', fontsize=14)
plt.tight_layout()
```

```
train, test = train_test_split(df, test_size=1/5,stratify=df.default,random_state=0)
```
